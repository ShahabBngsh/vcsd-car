```python
import sounddevice as sd
import librosa
from scipy.io.wavfile import write
from pydub import AudioSegment
import socket
import os
```

```python
import numpy as np
X_train=np.load("trainTestSplitLists/X_train_25_5backgroundPostLoud.npy")
Y_train=np.load("trainTestSplitLists/Y_train_25_5backgroundPostLoud.npy")
X_test=np.load("trainTestSplitLists/X_test_25_5backgroundPostLoud.npy")
Y_test=np.load("trainTestSplitLists/Y_test_25_5backgroundPostLoud.npy")
```

```python
print(X_test[0].shape)
print(Y_test.shape)
```

```python
from tensorflow import keras
modelFileName="lstm5_PL_200_370.h5"
modelstm = keras.models.load_model("VoiceModule/models/"+modelFileName)

#         modelstm.summary()
#         modebilstm.summary()
# temp=[]
# temp.append(mfccs[0])

#         print(np.shape(temp))
#         print(temp[0].shape)
count=0
prediction_array=[]
actual_output_array=[]
for mfccs in X_test:    
#     print(mfccs.shape)
    temp=[]
    temp.append(mfccs)
    y_probLstm=modelstm.predict(np.array(temp))
    y_classesLstm = y_probLstm.argmax(axis=-1)
    print("Lstm Prediction: "+ str(y_classesLstm[0]) +"Actual Output"+str(list(Y_test[count]).index(1)))
    prediction_array.append(y_classesLstm[0])
    actual_output_array.append(list(Y_test[count]).index(1)) 
    count+=1
# y_probBiLstm=modebilstm.predict(X_test)
# y_classesBiLstm = y_probBiLstm.argmax(axis=-1)

# print("BiLstm Prediction: "+ str(y_classesBiLstm[0]))

```
