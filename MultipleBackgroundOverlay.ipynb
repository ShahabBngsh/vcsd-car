<h1>This notebook does folllowing things</h1>
<h2>1. Selecting audio files having length <= 2.5s</h2>
<h2>2. Overlaying each file with background noise recording in 5 differnt environments</h2>
<h2>3. So each file is converted into 5 file with 5 differnt backgorund noises</h2>

```python
import librosa
from pydub import AudioSegment
audioLength=2.5
filePathOriginal='VoiceData/commands/'
filesBasicPath='VoiceData/multipleBackground/'
audioFileLength=[]
audio_path_ahistaChalo=[]
audio_path_bayainMuro=[]
audio_path_chalo=[]
audio_path_changeLane=[]
audio_path_dayainMuro=[]
audio_path_laneBadlo=[]
audio_path_rukJao=[]
audio_path_speedDown=[]
audio_path_speedUp=[]
audio_path_start=[]
audio_path_stop=[]
audio_path_tezChalo=[]
audio_path_turnLeft=[]
audio_path_turnRight=[]
fypLabAudio = AudioSegment.from_wav('VoiceData/commands/backgroundVoices/fypLab.wav')
roomAudio = AudioSegment.from_wav('VoiceData/commands/backgroundVoices/room.wav')
backGroundNoisesList=['VoiceData/BackgroundVoices/fypLab.wav','VoiceData/BackgroundVoices/uniCafe.wav','VoiceData/BackgroundVoices/room.wav','VoiceData/BackgroundVoices/hamzaAcadmy.wav','VoiceData/BackgroundVoices/traffic1.wav']
numberOfCommands=85
ahistaChalo_count=0
bayainMuro_count=0
chalo_count=0
changeLane_count=0
dayainMuro_count=0
laneBadlo_count=0
rukJao_count=0
speedDown_count=0
speedUp_count=0
start_count=0
stop_count=0
tezChalo_count=0
turnLeft_count=0
turnRight_count=0


for i in range(1,numberOfCommands+1):
    if i!=39 and i!=66 and i!=83:
        
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/ahista chalo.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):
                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/ahista chalo.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"ahistaChalo"+'/ahista chalo'+str(ahistaChalo_count)+'.wav', format="wav")
                ahistaChalo_count+=1
    if i!=46 and i!=39:
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/bayain muro.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):
                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/bayain muro.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"bayainMuro"+'/bayain muro'+str(bayainMuro_count)+'.wav', format="wav")
                bayainMuro_count+=1
    if i!=46 and i!=39:    
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/chalo.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):
                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/chalo.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"chalo"+'/chalo'+str(chalo_count)+'.wav', format="wav")
                chalo_count+=1
    if i!=46 and i!=39:
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/dayain muro.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):
            
                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/dayain muro.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"dayainMuro"+'/dayain muro'+str(dayainMuro_count)+'.wav', format="wav")
                dayainMuro_count+=1
    if i!=27 and i!=62 and i!=39 and i!=74:        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/ruk jao.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):
                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/ruk jao.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"rukJao"+'/ruk jao'+str(rukJao_count)+'.wav', format="wav")
                rukJao_count+=1
    if i!=5 and i!=18 and i!=19 and i!=39 and i!=74 and i!=83:
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/tez chalo.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):
                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/tez chalo.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"tezChalo"+'/tez chalo'+str(tezChalo_count)+'.wav', format="wav")
                tezChalo_count+=1
    if i!=25 and i!=37 and i!=36 and i!=38 and i!=53 and i!=39:
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/lane badlo.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/lane badlo.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"laneBadlo"+'/lane badlo'+str(laneBadlo_count)+'.wav', format="wav")
                laneBadlo_count+=1
    if i!=36 and i!=38:  
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/change lane.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/change lane.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"changeLane"+'/change lane'+str(changeLane_count)+'.wav', format="wav")
                changeLane_count+=1
    if i!=46 and i!=51:
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/speed down.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/speed down.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"speedDown"+'/speed down'+str(speedDown_count)+'.wav', format="wav")
                speedDown_count+=1
    if i!=26:
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/speed up.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/speed up.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"speedUp"+'/speed up'+str(speedUp_count)+'.wav', format="wav")
                speedUp_count+=1
    if i!=10 and i!=38 and i!=39 and i!=43 and i!=44 and i!=55:        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/start.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/start.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"start"+'/start'+str(start_count)+'.wav', format="wav")        
                start_count+=1
    if i!=10 and i!=36 and i!=43:
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/stop.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/stop.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"stop"+'/stop'+str(stop_count)+'.wav', format="wav")
                stop_count+=1
    if i!=36 and i!=38:
        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/turn left.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/turn left.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"turnLeft"+'/turn left'+str(turnLeft_count)+'.wav', format="wav")
                turnLeft_count+=1

    if i!=38 and i!=10 and i!=74:        
        if librosa.get_duration(filename=filePathOriginal+str(i)+'/turn right.wav')<=audioLength:
            for backgroundNoises in range(0,len(backGroundNoisesList)):

                sound1 = AudioSegment.from_wav(filePathOriginal+str(i)+'/turn right.wav')
                combined_sounds = AudioSegment.from_wav(backGroundNoisesList[backgroundNoises]).overlay(sound1)
                combined_sounds.export(filesBasicPath+"turnRight"+'/turn right'+str(turnRight_count)+'.wav', format="wav")
                turnRight_count+=1
print("Loading is completed Done")
```
